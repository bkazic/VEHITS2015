# Traffic Flow Prediction (using Loop Counter Sensor Data)

Due  to  increasing  demand  and  growing  cities,  traffic  prediction  has  been  a  topic  of  interest  for  many 
researchers  for  the  past  few  decades.  The  availability  of  large  amounts  of  traffic-related  data  and  the 
emerging field of  machine learning algorithms has led to a significant leap towards data-driven  methods. In 
this  paper,  loop counter data  are used  to  develop  models that  can  predict traffic  flow  for  several  different 
prediction intervals  into the future.  In depth exploratory data analysis and statistical testing is performed to 
obtain  good  quality  informative  features.  Several  feature  sets  were  compared  by  using  different  machine 
learning methods:  Ridge Regression, SVR and Random  Forests.  The results show that  in order to  obtain 
good prediction results  thorough  feature extraction is  just as  or even  more important than  learning  method 
selection.  Good  features  enables us to use  less complex methods, which run  faster, are more reliable and 
easier  to  maintain.  In  conclusion,  we  address  ideas  regarding  how  predictions  could  be  improved  even 
further.

# 2. Related Work

In  general,  traffic  prediction  studies  can  be 
categorized  into  three  major  categories:  naïve 
methods,  parametric  methods  and  non-parametric 
methods  (Van Lint & Van Hinsbergen, 2012). Naïve 
methods  are  usually  simple  non-model  baseline 
predictors, which can sometimes return good results. 
Parametric  models  are  based  on  traffic  flow  theory 
and are researched separately and in parallel to nonparametric,  more  data-driven  machine  learning 
methods.

<img src="files/Final/Figures/taxonomy.png" width="50%" height="50%" />

A  strong  movement  towards  non-parametric 
methods  can  be  observed  in  the  recent  years, 
probably  because  of  increased  data  availability,  the 
progress  of  computational  power  and  the 
development  of  more  sophisticated  algorithms 
(Vlahogianni et al., 2014). Non-parametric  does not 
mean  models  are  without  parameters;  but  refers  to 
model’s parameters, which are flexible and not fixed 
in advance. The model’s structure as well as model 
parameters  are  derived  from  data.  One  significant 
advantage  of  this  approach  is  that  less  domain 
knowledge  is  required  in  comparison  to  parametric 
methods, but also more data is required to determine 
a  model.  This  also  implies  that  successful 
implementation  of  data-driven  models  is  highly 
correlated to the quality of available data.

Non-parametric  methods  can  be  further 
subdivided  into  two  subgroups:  classical  statistical 
regression  approaches  and  data-driven  machine 
learning  approaches  (Karlaftis  &  Vlahogianni, 
2011).  From  the  group  of  statistical  methods,  the 
local linear regression algorithm  yields  surprisingly 
good  results,  especially  on  highway  data  (Rice  & 
van  Zwet,  2004).  In  contrast,  traffic  in  urban  areas 
can be  much  more dynamic  and non-linear,  mainly 
because  of  the  presence  of  many  intersections  and 
traffic  signs.  In  such  environments,  data -driven 
machine-learning  approaches,  such  as  neural 
networks  (Van  Hinsenberg  et  al.,  2007)  and  SVR 
(Vanajakshi  &  Rilett,  2007),  can  be  more 
appropriate,  due  to  their  ability  to  model  highly 
nonlinear  relationships  and  dynamic  processes.  In 
this  research,  we  test  methods  from  both  groups: 
statistical  regression  approaches  and  more  complex 
nonlinear algorithm.

From the previously mentioned database, we have created a dump file (in csv format) of one specific traffic loop sensor (*SensorID: 0178_11*). This sensor is placed on one of the busiest roads, located on the bypass of the Slovenias capital city Ljubljana. This 6 months long historical data dump will be used for further analysis. - **speed:** average speed (*km/h*) of vehicles, per hour. Speed of every vehicle is almost impossible to track and is therefore estimated from other parameters.
- **traffic status:** Numeric status of the traffic. 1 being *"normal traffic"*, and 5 being *"heavy traffic with congestions"*.

## 3.2 Data preprocessing

With a basic sanity check we can see some interesting findings. For example, average speed is around 97*km/h*, which make sense, since speed limit on this section is 100*km/h*. We can also see that number of samples is the same for all parameters. As we will see later, this does not mean that we do not have any missing data, but it just means that when we have missing data, all the parameters are missing. We can also spot some irregular values. For example we know that "TrafficStatus" can have values only between 1 and 5, but in the data bellow, we can see that minimum value is 0. In the next sections, we will try to handle this corrupt data and missing values.

```python
traffic_data.describe()
```

### Corrupt data

If we check look for the records where "TrafficStatus" parameter equals to zero, we can quickly see that this records are corrupted. We can see that for this records, all parameters equals to zero. This does not make sense. For example, if "Speed" would actually be zero due to some heavy traffic congestion or accident, than "Occupancy" parameter shouldn't be zero, but something big. Also from the DateTime parameter, we can see that something is wrong, since the date and time is repeated. We can conclude that all the records where all the parameters are 0 are corrupted, therefore we will mark them as NaN (null) and deal with them as missing values (in the next section). 

```python
traffic_data[traffic_data["TrafficStatus"]==0].head()
```

```python
traffic_data[traffic_data["Speed"]==0] = np.nan
```

### Duplicated values are Missing values

Another obvious irregularity can be seen from the DateTime column, where we can see some duplicated records. As it turns out, this are actually the missing values, but since the web API is not updated when the data are not streaming this creates duplicate values in the historical dump. We have to detect this duplicates and mark them as NaN (null) values.

```python
# Delete all duplicated records
traffic_data.drop_duplicates(inplace=True)

# Find missing gaps, and mark them as NaN.
idx = pd.date_range(start=traffic_data.index.min(),end=traffic_data.index.max(), freq="5min")
traffic_data = traffic_data.reindex(idx)
traffic_data[17:23]
```

Now that all corrupted data and missing data have value Nan, we can decide what to do with missing data. For now we will just delete them, because we still have a lot of other records. It is not that trivial to replace missing values if they missing for a whole day or more. Previous value imputation, or some linear interpolation would not be a good idea in this case. We would probably need to use something like linear regression to replace missing values.

```python
traffic_data.dropna(inplace=True)
traffic_data[17:23]
```

### Adding a "Target"

For the reasons that was discussed in previous chapter "Flow" parameter was selected as the target parameter. "Target" column is added to the dataframe, by duplicating the "Flow" and shifting it for the same period as we want our predictions. In this work, we have chosen 5h. This column will be used to learn the model how to predict traffic flow for one hour into the future.

```python
# Create new column "Target" by shifting colum "Flow"
traffic_data["Target"] = traffic_data["Flow"].shift(-5, freq='H')

# Again all NaN values, that were caused by missing values
traffic_data.dropna(inplace=True)
traffic_data.head()
```

By performing sanity check again, we can see that we have reduced our dataset by 37% (from 46447 to 29215) with cleaning, but we still have reasonable amount of data to train and test our prediction model.  Overall, the data looks better now, for example minimum row does not contain zeros any more. We can also notice that the Speed mean slightly increased (from 96.8 *km/h* to 97.3 *km/h*), due to deleted corrupted zero values. We now have a clean dataset on which we can perform some statistic tests and train our prediction model.

```python
traffic_data.describe()
```

## 3.3 Exploratory Data Analysis

### Is traffic during the week any different?

In the previous chapter, it was determined that traffic
during the weekend is significantly different than on 
weekdays, which is intuitively understood to be true. 
This  implies  that  it  would  be  useful  to  include  this 
information  (whether  it  is  weekend  or  weekday)  in 
the form of a new feature, when training our   model. 
What  about  traffic  during  the  days  in  the  week? 
Would it also be useful to have the day of the week 
as a feature? 

In order to answer this question, we performed a 
statistical  T-test,  with  the  assumption  (null 
hypothesis) that traffic flow during weekdays is the 
same.  Since  the  traffic  flow  parameter  is  nonnormally  distributed,  we  performed  Man-Whitney 
U-test  in  order  to  test  our  null  hypothesis.  The  pvalue, which determines statistical significance of a 
hypothesis, was computed between all combinations 
of days in the  week.  A small p-value (smaller than 
the  critical  value,  usually  0.05)  means  that  we  can 
reject the null hypothesis; otherwise we cannot reject 
it.

```python
days = ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"]
p_values = pd.DataFrame(index=days, columns=days)

for i_idx, i_day in enumerate(days):
    for j_idx, j_day in enumerate(days):
        days1 = traffic_data["Flow"][traffic_data.index.dayofweek == i_idx]
        days2 = traffic_data["Flow"][traffic_data.index.dayofweek == j_idx]
        p = stats.mannwhitneyu(days1, days2)[1]*2
        p_values.loc[i_day][j_day] = round(p, 4)

#p_values
```

The  results  are  presented  in  the  form  of  a  heat 
map.  We  can  see  that  the  only 
combinations  for  which  we  cannot  reject  our  null 
hypothesis  (p-value  >  0.05)  are  the  days  from 
Tuesday  to  Thursday.  However,  we  can  reject  our 
null  hypothesis  for  all  other  days,  meaning  that 
traffic  is  significantly  different  for  these  days. 
According  to  these  results,  it  would  make  sense  to 
also  include  the  day  of  the  week  as  a  feature  in  to 
our model.

```python
# Plot heatmap of p-test results
# Ref: http://stackoverflow.com/questions/20520246/create-heatmap-using-pandas-timeseries
from matplotlib import patheffects
fig = plt.figure(figsize=(6, 6), dpi=100)
plt.title("P values results from Man-Whitney statistical U-test", y=1.2)
m, n = p_values.shape
ax = plt.imshow(p_values.astype(float), interpolation='nearest', cmap='Oranges').get_axes()
ax.xaxis.tick_top()
ax.set_xticks(np.linspace(0, n-1, n))
ax.set_xticklabels(p_values.columns, rotation=90)
ax.set_yticks(np.linspace(0, m-1, m))
ax.set_yticklabels(p_values.index)
ax.grid('off')

# Display also values
for i in range(m):
    for j in range(n):
        ax.text(j, i, '{:.2f}'.format(p_values.iget_value(i, j)),
                size='medium', ha='center', va='center',
                path_effects=[patheffects.withSimplePatchShadow(shadow_rgbFace=(1,1,1))])
```

### Feature Correlations

We have allready got a little intuition about our data with visualizing different traffic parameters over one week (figure in 2.1 Data descriton section). Another good view to get an even better feel for our data is, by displaying correlations between fields (parameters).

One of the assumptions of many linear models is that input features are statistically independent (uncorrelated) with each other. Some methods (e.g. generalized least squares) are capable of handling correlated errors, although they typically require significantly more data, unless some sort of regularization is used to bias the model towards assuming uncorrelated errors. Therefore, it is necessary to choose an appropriate set of features in order to obtain the best results from our models. 

In order to find if we have any strongly correlated features, we will visualize correlations between features and target value. 

```python
from pandas.tools.plotting import scatter_matrix
_ = scatter_matrix(traffic_data[["Flow","Gap","Occupancy","Speed","AvrFlow","Target"]], 
                   alpha=0.2, figsize=(10,10), diagonal='kde')
```

From the results above, we can see that we have a strong linear correlation between “Flow”, “Occupancy” and “TrafficStatus”, while “Gap” and “Flow” seems to be non-linearly (pollynomically) correlated. In one way, it makes sense that all these parameters are correlated, since they are all measuring the same state. However, these correlated features do not bring any new information to the model; even worse, they can induce noise and corrupt our prediction models.

What we can also observe is that the traffic measurements are not correlated with the “Target” value, which may indicate that additional features will have to be constructed in order for the (linear) models to work well.


## 3.3 Feature Engineering

It is a well-known fact that feature engineering is  a
key to success  in applied machine learning. Feature 
engineering  is the process of transforming raw data, 
into  features  that  better  represent  the  underlying 
problem  to  the  predictive  models.  Better  the 
features,  better  the  results.  The  best  way  to  extract 
good  representative  features,  is  by  exploring  and 
understanding,  which  is  what  we  did  in  previous 
section.

We  have  already  discovered  that  there  exists  a 
relationship  between  datetime  and  other  attributes. 
Since  it  can  be  difficult  for  a  model  to  take 
advantage  of  datetime  native  form  (2014-01-01 
01:00:00),  we  want  to  decompose  a  datetime  into 
consistent  parts,  that  may  be  more  informative  for 
the model. We introduce three new features:

- hour of day *[0 -23]*
- day of week *[0 - 6]*
- weekday/weekend *[0 - 1]*

```python
traffic_data["HourOfDay"] = traffic_data.index.hour # Could try cyclic hours
traffic_data["DayOfWeek"] = traffic_data.index.dayofweek
traffic_data["Month"] = traffic_data.index.month # month of year [1-12]
traffic_data["Weekday"] = traffic_data.DayOfWeek.apply(lambda x: 1 if x < 5 else 0)

print 'df_train_dummy shape is ', traffic_data.shape
traffic_data.head()
```

### Adding Dummy features

From previous empirical experiences, it can also 
be  beneficial  for  some  models,  to  convert  these
ordinal  variables into dummy/indicator  (categorical)
variables  (i.e.  “DayOfWeek”  feature  3  is 
transformed to 0001000).

```python
traffic_dummy = traffic_data
traffic_dummy = traffic_dummy.join(pd.get_dummies(traffic_dummy['HourOfDay'], prefix='hour'))
traffic_dummy = traffic_dummy.join(pd.get_dummies(traffic_dummy['DayOfWeek'], prefix='day'))
traffic_dummy = traffic_dummy.join(pd.get_dummies(traffic_dummy['Month'], prefix='month'))
traffic_dummy = traffic_dummy.drop(['Month','DayOfWeek','HourOfDay'], axis = 1)
#df_enriched.head(10)
#df_enriched.dtypes
print 'df_train_dummy shape is ', traffic_dummy.shape
```

### Adding Localized Average feature

Another  important  information,  derived  from 
outcomes  of  previous  chapter,  is  weekly  traffic  is 
very similar. Therefore, traffic average (according to 
time and day of  week) can represent normal traffic 
surprisingly  well.  This  information  can  be  used  as 
feature,  that  informs  what  is  the  traffic  status  on 
average,  on  specific  time  in  the  future.  For  this 
purpuse we create extra feature:

- AvrFlow

### Creating Feature sets

Although  good  feature  engineering  is  very 
important  in order to achieve  high-quality prediction 
results, the actual success is a combination between 
the model that we choose, the data, and the extracted 
features  that  we  use.  In  order  to  obtain  the  best 
combination,  4  different  datasets  were  created.

Our datasets is also splited into 2 subset: training model and testing model. On the first subset, we will train the models and on the other, models will be tested. If we train and test our model on the same dataset, we risk the problem of over fitting (also this can be considered as cheating).

```python
# Unshuffled split
splitting_point = int(0.8 * traffic_data.shape[0])
traffic_data_train = traffic_data.iloc[:splitting_point]
traffic_data_test = traffic_data.iloc[splitting_point:]

# Feature set 1: "only_measurements"
only_measurements = traffic_data[['Flow','Gap','Occupancy','Speed','TrafficStatus']]
only_measurements_train = only_measurements.iloc[:splitting_point]
only_measurements_test = only_measurements.iloc[splitting_point:]
print 'Features set 1: only_measurements shape is ', only_measurements.shape 

# Feature set 2: "with_datetime"
with_datetime = traffic_data.drop(['SensorID', 'AvrFlow', 'Target'], axis = 1)
with_datetime_train = with_datetime.iloc[:splitting_point]
with_datetime_test = with_datetime.iloc[splitting_point:]
print 'Features set 2: all_features shape is ', with_datetime.shape 

# Feature set 3: "dummy_datetime"
dummy_datetime = traffic_dummy.drop(['SensorID', 'AvrFlow', 'Target'], axis = 1)
dummy_datetime_train = dummy_datetime.iloc[:splitting_point]
dummy_datetime_test = dummy_datetime.iloc[splitting_point:]
print 'Features set 3: dummy_datetime shape is ', dummy_datetime.shape 

# Feature set 4: "with_avr"
with_avr = traffic_data.drop(['SensorID', 'Target'], axis = 1)
with_avr_train = with_avr.iloc[:splitting_point]
with_avr_test = with_avr.iloc[splitting_point:]
print 'Features set 4: with_avr shape is ', with_avr.shape 

# Feature set 5: "with_avr_dummy"
with_avr_dummy = traffic_dummy.drop(['SensorID', 'Target'], axis = 1)
with_avr_dummy_train = with_avr_dummy.iloc[:splitting_point]
with_avr_dummy_test = with_avr_dummy.iloc[splitting_point:]
print 'Features set 5: with_avr_dummy shape is ', with_avr_dummy.shape 
```

We will also need target values.

```python
# Target values
traffic_target = traffic_data['Target']
traffic_target_train = traffic_target.iloc[:splitting_point]
traffic_target_test = traffic_target.iloc[splitting_point:]
print 'Target values: traffic_target shape is ', traffic_target.shape  
```

### Learning Curve

```python
# Ref: http://scikit-learn.org/stable/auto_examples/plot_learning_curve.html
def plot_learning_curve(ax, estimator, X, y, title=None, xlabel=None, ylabel=None, ylim=None, cv=None,
                        n_jobs=1, train_sizes=np.linspace(.1, 1.0, 5)):
    
    ax.set_title(title)
    if ylim is not None:
        ax.set_ylim(*ylim)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    train_sizes, train_scores, test_scores = learning_curve(estimator, X, y, cv=cv, n_jobs=n_jobs, train_sizes=train_sizes)
    train_scores_mean = np.mean(train_scores, axis=1)
    train_scores_std = np.std(train_scores, axis=1)
    test_scores_mean = np.mean(test_scores, axis=1)
    test_scores_std = np.std(test_scores, axis=1)
    ax.grid()

    ax.fill_between(train_sizes, train_scores_mean - train_scores_std, train_scores_mean + train_scores_std, alpha=0.1, color="r")
    ax.fill_between(train_sizes, test_scores_mean - test_scores_std, test_scores_mean + test_scores_std, alpha=0.1, color="g")
    ax.plot(train_sizes, train_scores_mean, 'o-', color="r", label="Training score")
    ax.plot(train_sizes, test_scores_mean, 'o-', color="g", label="Cross-validation score")
    #ax.legend(loc=4)
    
    return ax
```

### Parameter Optimisation (grid search)

```python
# Ref: http://scikit-learn.org/stable/auto_examples/grid_search_digits.html
X = with_avr_train
y = traffic_target_train

X_train, X_test, y_train, y_test = cross_validation.train_test_split(X, y, test_size=0.2, random_state=0)

tuned_parameters = [{'n_estimators': [10, 100, 500],
                    'min_samples_leaf': [1, 10, 20]}]

#error_measures = ['r2', 'mean_squared_error', 'mean_absolute_error']
error_measures = ['r2']

#gs_results = pd.DataFrame(index = n_estimators)
#gs_results = pd.DataFrame(index=pd.MultiIndex.from_arrays([[],[]], names=['error measure', 'n estimators']))

for error_measure in error_measures:
    model = GridSearchCV(RandomForestRegressor(), tuned_parameters, cv=5, scoring=error_measure, n_jobs=8)
    model.fit(X_train, y_train)
    
    print("\n== Tuning hyper-parameters for: %s ==\n" % error_measure)

    for params, mean_score, scores in model.grid_scores_:
        print("%0.3f (+/-%0.03f) for %r" % (mean_score, scores.std() / 2, params))
        #gs_results.loc[(error_measure, params['n_estimators']), params['max_features']] = mean_score
        
    print("\nBest parameters set found on development set:\n")
    #best_estimator_ returns the best estimator chosen by the search
    print(model.best_estimator_)
```

experience  the  largest  errors  with  anomalies.  By 
using  additional  data  sources  that  could  describe 
such anomalous traffic, we could probably increase 
our  prediction  accuracy.  Examples  of  such 
additional data  sources  would be information about 
holidays,  weather  prediction  reports,  traffic  status 
reports, etc.

# Conclusion

In this research, we have compared the perfor mance 
of  several  machine  learning  methods  used  for 
predicting traffic flow. Feature engineering was also 
described  and  recognized  as  a  key  component  for 
good results.

The results show that simple naïve methods, such 
as historical average, are surprisingly effective when 
making  long-term  predictions  (more  than  one  hour 
into  the  future),  while  using  current  traffic 
measurements as naïve method for prediction works 
well when making more short term predictions (less 
than  1h).  In  one  sense  this  is  to  be  expected ,   but 
what is more important is, that results shows, that by 
using  machine  learning  methods  which  includes 
both;  historical  averages  and  current  values, 
predictions  are  better  than  both  previously 
mentioned  naïve  predictors,  for  all  prediction 
horizons.

Also noteworthy is the fact that by constructing 
high quality features, simple methods, such as linear 
regression can work as well or even better than other 
more  sophisticated  algorithms  (such  as  random 
forests,  SVR,  etc.).  By  using  less  complex  models, 
optimal  model  parameters  are  found  more  readily,

# Future Work


We also state that the main disadvantage of such 
model  is  its  inability  to  predict  unusual  traffic, 
usually  caused  by  other  complex  events,  such  as 
traffic accidents  nearby, bad weather, holidays, etc. 
This  would,  in  fact,  also  be  the  most  desirable  and 
useful  information,  since  the  frequent  commuter  is 
aware of normal traffic state but is more interested in 
unusual  traffic  events.  The  main  reason  for  this 
disadvantage  is  that  model  uses  only  historical 
traffic data. We believe that by including additional 
data  sources  (such  as  weather  forecasts,  traffic 
related information, special days statuses, etc.) in the 
model, prediction could be significantly improved. 

Therefore,  our  future  plan  is  to  collect  several 
quality traffic related data sources and combine them 
in order to generate better traffic prediction models. 
We  intend  to  test  different  data  fusion  approaches
such as: 
- *early (or full) integration:* which transforms 
data  sources  into  a  single  feature-based  table,  
- *late (decision) integration:*  where each data  source give 
rise  to  a  separate  model  and  predictions  are  later 
fused, 
- *intermediate  (partial)  integration:*  where  data  are
fused  through  inference  of  a  single  joint  model. Very  recent  algorithms  (providing  very 
good  results  in  the  field  of  bioinformatics) (Žitnik & Zupan, 2013).
